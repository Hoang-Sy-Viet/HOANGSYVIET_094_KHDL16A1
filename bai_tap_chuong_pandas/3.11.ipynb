```python
import numpy as np
import pandas as pd
```

```python
# a:
stocks1 = pd.read_csv('stocks1.csv')
print(stocks1.head(5))
print(stocks1.tail())
print(stocks1.dtypes)
print(stocks1.info())
```

```python
# b
stocks2 = pd.read_csv('stocks2.csv')
print(stocks2.head(5))
print(stocks2.tail())
print(stocks2.dtypes)
print(stocks2.info())
```

```python
# c:
# b
companies = pd.read_csv('companies.csv')
print(companies.head())
print(companies.dtypes)
print(companies.info())
```

```python
# 2: Cho biết trong stocks1 có dữ liệu Null hay không?
print(stocks1.isnull().any())


stocks1['high'] = stocks1.groupby('symbol')['high'].transform(lambda x: x.fillna(x.max()))
print()
stocks1['low'] = stocks1.groupby('symbol')['low'].transform(lambda x: x.fillna(x.min()))
print()
print(stocks1.isnull().any())
print(stocks1)
```

```python
# 3:
stocks = pd.concat([stocks1,stocks2],ignore_index=True)
stocks.tail(15)
```

```python
# 4:
stocks_companies = pd.merge(stocks,companies,left_on='symbol', right_on='name')
stocks_companies.head()
```

```python
# 5:
y_5 = stocks_companies.groupby('name')['open','high','low','close','volume'].mean()
y_5
```

```python
# 6:
y_6 = stocks_companies.groupby('name')['close'].agg(['mean','max','min'])
y_6
```

```python
# 7:
stocks_companies['parsed_time'] = pd.to_datetime(stocks_companies['date'])
print(stocks_companies['parsed_time'].dtype)
```

```python
# 8:
stocks_companies['result'] = ''   
stocks_companies['result'] = stocks_companies.apply(lambda x: 'up' if x['close']>x['open'] else 'down',axis=1)
stocks_companies
```
