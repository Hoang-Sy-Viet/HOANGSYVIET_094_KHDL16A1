```python
import pandas as pd
import numpy as np
```

```python
drink = pd.read_csv('drinks.csv',index_col=0)
print(drink)
```

```python
# 1:
# 1.1:
print(drink.dtypes)
# 1.2:
drink.shape
# 1.3:
print(drink.columns)

# 1.4:m 5 dòng dữ liệu đầu tiên (head) và cuối cùng (tail) của drink.
print("5 dong dau tien:\n",drink.head(5))
print("5 dong cuoi:\n",drink.tail())
```

```python
# 2. Cho biết số lượng bia tiêu thụ trung bình ở mỗi châu lục.
y_2 = drink.groupby(['continent'])['beer_servings'].mean()
y_2
```

```python
# 3. Cho biết thông tin thống kê tổng quát (describe) số lượng rượu vang được tiêu thụở mỗi châu lục.
y_3 = drink.groupby('continent')['wine_servings'].describe()
y_3
```

```python
# 4. Cho biết số lượng các loại bia và rượu tiêu thụ trung bình (mean) ở mỗi châu lục
y_4 = drink.groupby('continent')['wine_servings','beer_servings', 'spirit_servings'].mean()
y_4
```

```python
# 5. Cho biết giá trị trung vị (median) cho các loại bia và rượu tiêu thụ ở mỗi châu lục.
y_5 = drink.groupby('continent')['wine_servings','beer_servings', 'spirit_servings'].median()
y_5
```

```python
# 6. Cho biết số lượng rượu mạnh (spirit_servings) tiêu thụ trung bình, lớn nhất và nhỏ nhất ở mỗi châu lục
y_6 = drink.groupby('continent')['spirit_servings'].agg(['mean','max','mean'])
y_6
```

```python
# 7. Sắp xếp dữ liệu tăng dần (sort_values) theo số lượng bia tiêu thụ.
drink_sort = drink.sort_values(by='beer_servings')
df = drink_sort.loc[:,['country','beer_servings']]
print(df)
print()
print(" 5 quốc gia có lượng tiêu thụ bia nhiều nhất :\n",df.iloc[:5,0])
# print(" 5 quốc gia có lượng tiêu thụ bia it nhất :\n",df.loc[[135,65,62,45,117],['country']])
print(" 5 quốc gia có lượng tiêu thụ bia it nhất :\n",df.tail()['country'])
```
